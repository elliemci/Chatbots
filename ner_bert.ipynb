```python
from transformers import pipeline
```

```python
# default model for "ner" task is BERT transformer;
# pass aggregation_strategy parameter for grouping
# together parts of the same entity in a single  object
ner_pipeline = pipeline(task='ner', aggregation_strategy="SIMPLE")
```

```python
text = "My name is Ellie and I live on Boulder Street"
ner_pipeline(text)
```

```python
# BERT-base-NER is a fine-tuned BERT ready to use for Named Entity Recognition
# it recognizes location=LOC, organizations=ORG, person=PER and Miscellaneous(MISC)
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)
ner_result = nlp(text)
print(ner_result)
```
